```python
x = np.array([3,11,16,23,27,48,55,56,67,78])
y = np.array([13,23,34,49,47,61,69,78,84,89])
```

### Bar Plot

```python
percentage = [89.34, 45.34, 76.45, 65.90, 11.34, 71.90]
subject = ['python', 'PHP', 'Data Structures','Java', 'C', 'OOP']
```

```python
plt.title('Result Analysis')
plt.xlabel('Subject')
plt.ylabel('% of Passing')
plt.grid()
plt.bar(subject, percentage, color='brown')
```

```python
plt.title('Result Analysis')
plt.xlabel('% of Passing')
plt.ylabel('Subject')
plt.grid()
plt.barh(subject, percentage, color='blue')
```

### Pie Chart

```python
enrollment = [167, 67, 89, 106, 48, 99]
courses = ['python', 'PHP', 'Data Structures','Java', 'C', 'OOP']
```

```python
plt.figure(figsize=(10,10))
plt.pie(enrollment, labels=courses, autopct='%2.2f%%');  ### autopct = percentage
```
